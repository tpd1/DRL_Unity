```python
import pandas
from packaging import version

import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
from scipy import stats
import tensorboard as tb
```

### Import the tensorboard experiment as a DataFrame

```python
df_ent = pd.read_csv("final_policy_entropy.csv")
df_ent
```

```python
df_ep_len = pd.read_csv("final_episode_length.csv")
df_ep_len
```

### Display Episode Length Graph

```python
plt.figure(figsize=(5, 4))

sns.lineplot(x=df_ep_len["Step"], y=df_ep_len["Value"], color='#34186c',
             ).set_title("Episode Length")

plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x/1e6:.1f}'))
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(base=1e6))
plt.gca().xaxis.set_minor_locator(ticker.AutoMinorLocator(n=2))

plt.xlim(xmax = 10.5e6, xmin = 0)
plt.ylim(ymax = 850, ymin = 0)
plt.xlabel("Environment Steps (millions)")
plt.ylabel("Episode Length (steps)")
```

### Display Entropy Loss Graph

```python
plt.figure(figsize=(5, 4))

sns.lineplot(x=df_ent["Step"], y=df_ent["Value"], color='#34186c',
             ).set_title("Policy Entropy")

plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x/1e6:.1f}'))
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(base=1e6))
plt.gca().xaxis.set_minor_locator(ticker.AutoMinorLocator(n=2))

plt.xlim(xmax = 10.5e6, xmin = 0)
plt.ylim(ymax = 2.25, ymin = 0)
plt.xlabel("Environment Steps (millions)")
plt.ylabel("Entropy Value")
```
